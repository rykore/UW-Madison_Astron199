```python
import matplotlib.pyplot as plt
import numpy as np

fig1,axes1 = plt.subplots(1,4, sharex ='col', sharey = 'row', 
                       gridspec_kw = {'hspace': 0, 'wspace': 0.076}, figsize = (20,5))

axes1 = axes1.flatten()

plot_list = ['6', '10', '14', '18']
mdust_list = ['3','4','5','6']

for i,name1 in enumerate(plot_list):
    file_name1 = '/Users/riyakore/project_astronomy199/Sz71_model/sed_href_{0}.txt'.format(name1)
    data1 = np.loadtxt(file_name1)
    
    axes1[i].loglog(data1[:,0],data1[:,1],'r')
    axes1[i].loglog(data1[:,0],data1[:,2],'grey')
    axes1[i].set_xlim(0.1,1e4)
    axes1[i].set_ylim(1e-16,1e-9)
    axes1[i].set_title('h_ref = {0}'.format(name1), x = 0.8, y = 0.875)
    
fig2,axes2 = plt.subplots(1,4, sharex ='col', sharey = 'row', 
                       gridspec_kw = {'hspace': 0, 'wspace': 0.076}, figsize = (20,5))

axes2 = axes2.flatten()

for j,name2 in enumerate(mdust_list):
    file_name2 = '/Users/riyakore/project_astronomy199/Sz71_model/sed_mdust_{0}.txt'.format(name2)
    data2 = np.loadtxt(file_name2)
    
    axes2[j].loglog(data2[:,0],data2[:,1],'r')
    axes2[j].loglog(data2[:,0],data2[:,2],'grey')
    axes2[j].set_xlim(0.1,1e4)
    axes2[j].set_ylim(1e-16,1e-9)
    axes2[j].set_title('m_dust = 10e-{0}'.format(name2),x = 0.75, y = 0.875)
    
axes1[0].set_xlabel('wavelength(um)',fontsize=12)
axes1[0].set_ylabel('$\nuf_\nu$ (erg/cm^2)',fontsize=12)

axes2[0].set_xlabel('wavelength(um)',fontsize=12)
axes2[0].set_ylabel('$\nuf_\nu$ (erg/cm^2)',fontsize=12) 
 
```
