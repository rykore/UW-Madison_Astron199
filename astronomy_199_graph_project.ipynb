```python
import matplotlib.pyplot as plt
import numpy as np

fig,axes = plt.subplots(4,5, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0}, figsize = (15,10))

axes = axes.flatten()

plot_row_1 = ['HTLup','GWLup','IMLup','RULup','Sz114']
plot_row_2 = ['Sz129', 'MYLup', 'HD142666', 'HD143006','AS205']
plot_row_3 = ['SR4', 'Elias20','DoAr25', 'Elias24', 'Elias27']
plot_row_4 = ['DoAr33', 'WSB52', 'WaOph6', 'AS209' , 'HD163296']

name_list = ['HTLup','GWLup','IMLup','RULup','Sz114','Sz129', 'MYLup', 'HD142666', 'HD143006','AS205','SR4', 'Elias20','DoAr25', 'Elias24', 'Elias27','DoAr33', 'WSB52', 'WaOph6', 'AS209' , 'HD163296']

for i,name in enumerate(name_list):
    file_name ='/Users/riyakore/project_astronomy199/{0}.SED.txt'.format(name)
    data = np.loadtxt(file_name, skiprows = 10, usecols = (0,1), dtype = float)
    f_nu = data[:,1]*1e-23 #Jy->erg/cm^2/s/Hz
    
    wavelength_um = data[:,0]
    nu = 3e8*1e6/wavelength_um
    nu_fnu = nu*f_nu #erg/cm^2/s
    
    axes[i].loglog(wavelength_um,nu_fnu,'bo')
    axes[i].set_xlim(0.1,1e4)
    axes[i].set_title(name, x = 0.8, y = 0.8)
    
    
axes[15].set_xlabel('wavelength(um)',fontsize=12)
axes[15].set_ylabel('$\nuf_\nu$ (erg/cm^2)',fontsize=12)
```
